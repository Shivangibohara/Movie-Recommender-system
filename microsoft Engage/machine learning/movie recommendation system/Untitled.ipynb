```python
import ast
#this module convert strings into list as with strings this convert function was ginving error
```

```python
def convert(text):
    L= []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L 
# we just want top 3 actors of the movies
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
# we just want director of the movie
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
#movies.head()
# we have converted overview from string to list so that we can merge it with other columns
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
#this function is used for removing spaces so that no two names can be treated same.like here sam washington is diff then sam mendes.
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
#movies['overview'] = movies['overview'].apply(collapse)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
#concatenated to make column tags
```

```python
new=movies[['movie_id','title','tags']]
```

```python
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
#here we have converted tags in a string
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
#converted whole string in lower case beacuse it is suggested to do so
```

```python
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
#this is done to convert all words to its root word like love is root word for loving loved love
```

```python
def stem(text):
    y = []
    
    for i in text.split():
      y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
#now we have to convert text to vector 
#so for that i used "bag of words" technique
#in this technique we will first combine all the tags and then will find top 5000 words which have high frequencies and then for each movie tag we will find the frequency of these 5000 words and will store in vector.
#then for recommendation we will find nearest vectors of that vector
```

```python
#from sklearn.feature_extraction.text import CountVectorizer
#cv = CountVectorizer(max_features=5000,stop_words='english')
#this is done to remove stop words like to,and for which are not required
#we have used sklearn library and used countvectorizer class
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
#print(vector.toarray())
```

```python
#as we are dealing with large datas so euclirean distance is not a good measure to deal with so we will calculate cosine distance that is angle between two points more then angle less similarities
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Twilight')
```
